```python
# импорт
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import matplotlib.cm as cm
from scipy.constants import G, pi

# вводные данные для расчетов
# реальные параметры Юпитера (от NASA 2024), хотя можно было поточнее, но это то, с чем я хотела работать для наглядности
R_jupiter = 71492  # радиус экватора км
M_jupiter = 1.898e27  # масса кг
T_rotation = 9.925  # период вращения часов
T_orbit = 4333  # период обращения дней
a_orbit = 7.785e8  # большая полуось км
e_orbit = 0.0489  # эксцентриситет
g_surface = 24.79  # ускорение свободного падения м/с в ^2
temp_cloud = -145  # температура облаков 
pressure_surface = 100000  # давление на "поверхности" Па
```

```python
# сравнение размеров планет
planets = ['Меркурий', 'Венера', 'Земля', 'Марс', 'Юпитер', 'Сатурн', 'Уран', 'Нептун']
radii = [2440, 6052, 6371, 3390, 71492, 60268, 25559, 24764]  # радиусы в км

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# круговая диаграмма относительных размеров
ax1.pie(radii, labels=planets, autopct='%1.1f%%', startangle=90)
ax1.set_title('Относительные размеры планет')

# гистограмма
colors = ['gray', 'orange', 'blue', 'red', 'brown', 'yellow', 'cyan', 'navy']
bars = ax2.bar(planets, radii, color=colors)
ax2.set_ylabel('Радиус (км)')
ax2.set_title('Абсолютные размеры планет')
ax2.tick_params(axis='x', rotation=45)

# нашу планету Землю для сравнения
earth_rad = radii[2]
for i, (planet, r) in enumerate(zip(planets, radii)):
    if r > earth_rad:
        earth_ratio = r / earth_rad
        ax2.text(i, r, f'x{earth_ratio:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# моделирование орбиты Юпитера
# симуляция движения по эллиптической орбите (для полного анализа, конечно, нужно учитывать все орбиты солнечной системы)
n_points = 365
theta = np.linspace(0, 2*np.pi, n_points)

# параметры орбиты (здесь упрощенная модель)
r_orbit = a_orbit * (1 - e_orbit**2) / (1 + e_orbit * np.cos(theta))

# создание сетки для гравитационного поля
x = np.linspace(-8e8, 8e8, 50)
y = np.linspace(-8e8, 8e8, 50)
X, Y = np.meshgrid(x, y)

# расчет гравитационного потенциала
R = np.sqrt(X**2 + Y**2)
V = -G * M_jupiter / (R + 1e6)  # добавляем малое число чтобы избежать деления на 0

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# орбита Юпитера
ax1.plot(r_orbit * np.cos(theta) / 1e8, r_orbit * np.sin(theta) / 1e8, 'b-', alpha=0.7)
ax1.scatter([0], [0], s=300, color='yellow', label='Солнце')
jupiter_pos = (a_orbit * (1 - e_orbit), 0)
ax1.scatter([jupiter_pos[0]/1e8], [0], s=100, color='brown', label='Юпитер (перигелий)')
ax1.set_xlabel('Расстояние (×10^8 км)')
ax1.set_ylabel('Расстояние (×10^8 км)')
ax1.set_title('Орбита Юпитера')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.axis('equal')

# гравитационное поле Юпитера
contour = ax2.contourf(X/1e8, Y/1e8, V, 20, cmap='viridis')
ax2.set_xlabel('Расстояние (×10^8 км)')
ax2.set_ylabel('Расстояние (×10^8 км)')
ax2.set_title('Гравитационный потенциал Юпитера')
plt.colorbar(contour, ax=ax2, label='Потенциал (Дж/кг)')

plt.tight_layout()
plt.show()
```

```python
# модель внутреннего строения (ну очень приблизительно)
layers = ['Водородная атмосфера', 'Металлический водород', 'Ядро']
depths = [0, 20000, 60000, 71492]  # глубины в км
pressures = [100, 2000000, 45000000, 100000000]  # давление в барах
temperatures = [-145, 6000, 20000, 35000]  # температура в °C

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# давление
ax1.plot(pressures, depths)
ax1.invert_yaxis()
ax1.set_xlabel('Давление (бар)')
ax1.set_ylabel('Глубина (км)')
ax1.set_title('Распределение давления')
ax1.grid(True, alpha=0.3)

# температура
ax2.plot(temperatures, depths)
ax2.invert_yaxis()
ax2.set_xlabel('Температура (°C)')
ax2.set_ylabel('Глубина (км)')
ax2.set_title('Распределение температуры')
ax2.grid(True, alpha=0.3)

# визуализация слоев
for i in range(len(layers)):
    depth_range = depths[i+1] - depths[i]
    circle = Circle((0, 0), depths[i+1]/1000, fill=False, linewidth=2, 
                   label=layers[i] if i < len(layers) else '')
    ax3.add_patch(circle)
    # подписи
    label_r = (depths[i] + depths[i+1]) / 2000
    ax3.text(0, -label_r, layers[i], ha='center', va='center')

ax3.set_xlim(-80, 80)
ax3.set_ylim(-80, 80)
ax3.set_aspect('equal')
ax3.set_title('Слои Юпитера')
ax3.axis('off')

plt.tight_layout()
plt.show()
```

```python
# анализ спутников Юпитера
# данные галилеевых спутников
moons = ['Ио', 'Европа', 'Ганимед', 'Каллисто']
orbit_radius = [421.6, 670.9, 1070.4, 1882.7]  # тыс. км
period = [1.77, 3.55, 7.16, 16.69]  # дни
radius = [1821.6, 1560.8, 2634.1, 2410.3]  # км
mass_relative = [1, 0.65, 2.02, 1.08]  # относительная масса (Ио = 1)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

# орбиты
angles = np.linspace(0, 2*np.pi, 100)
for i, (moon, r) in enumerate(zip(moons, orbit_radius)):
    x = r * np.cos(angles)
    y = r * np.sin(angles)
    ax1.plot(x, y, label=moon)
    
# Юпитер в центре
jupiter = Circle((0, 0), 50, color='orange', alpha=0.7)
ax1.add_patch(jupiter)
ax1.set_xlabel('Расстояние (тыс. км)')
ax1.set_ylabel('Расстояние (тыс. км)')
ax1.set_title('Орбиты галилеевых спутников')
ax1.legend()
ax1.axis('equal')

# период vs радиус орбиты
ax2.scatter(orbit_radius, period, s=100)
for i, (moon, x, y) in enumerate(zip(moons, orbit_radius, period)):
    ax2.text(x, y, moon, fontsize=9, ha='left')
ax2.set_xlabel('Радиус орбиты (тыс. км)')
ax2.set_ylabel('Период (дни)')
ax2.set_title('Закон Кеплера для спутников')
ax2.grid(True, alpha=0.3)

# размеры спутников
bars = ax3.bar(moons, radius, color=['red', 'blue', 'green', 'purple'])
ax3.set_ylabel('Радиус (км)')
ax3.set_title('Размеры спутников')
# добавим Землю для сравнения
ax3.axhline(y=6371, color='gray', linestyle='--', alpha=0.5, label='Земля (для масштаба)')
ax3.legend()

# фазы спутников (условная модель)
time = np.linspace(0, 30, 1000)
for i, moon_period in enumerate(period):
    phase = (np.sin(2 * np.pi * time / moon_period) + 1) / 2
    ax4.plot(time, phase + i*0.3, label=moons[i])
    
ax4.set_xlabel('Время (дни)')
ax4.set_ylabel('Фаза (условные единицы)')
ax4.set_title('Изменение фаз спутников')
ax4.legend(loc='upper right')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# атмосферные модели
# модель атмосферы - температура и давление по высоте
heights = np.linspace(-500, 500, 1000)  # км от "поверхности"
# упрощенная модель атмосферы
temperatures_atm = -145 + 2 * heights  # линейное приближение
pressures_atm = 100000 * np.exp(-heights / 50)  # экспоненциальный спад

# облачные слои
cloud_layers = [-50, 0, 50, 100]
cloud_names = ['Водяной лед', 'Гидросульфид аммония', 'Аммиак', 'Верхняя дымка']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# температура
ax1.plot(temperatures_atm, heights, 'r-', linewidth=2)
ax1.set_xlabel('Температура (°C)')
ax1.set_ylabel('Высота (км)')
ax1.set_title('Температурный профиль атмосферы')
ax1.grid(True, alpha=0.3)
# отметьте облачные слои
for h, name in zip(cloud_layers, cloud_names):
    ax1.axhline(y=h, color='gray', linestyle='--', alpha=0.5)
    ax1.text(-100, h, name, fontsize=8, va='center')

# давление
ax2.plot(pressures_atm, heights, 'b-', linewidth=2)
ax2.set_xlabel('Давление (Па)')
ax2.set_ylabel('Высота (км)')
ax2.set_title('Барический профиль атмосферы')
ax2.grid(True, alpha=0.3)
ax2.set_xscale('log')

# Большое Красное Пятно
# добавим врезку для вихря
ax_inset = ax2.inset_axes([0.6, 0.6, 0.35, 0.35])
# простая модель вихря
x_vortex = np.linspace(-10, 10, 100)
y_vortex = np.linspace(-10, 10, 100)
Xv, Yv = np.meshgrid(x_vortex, y_vortex)
vortex = np.exp(-(Xv**2 + Yv**2) / 20) * np.sin(np.sqrt(Xv**2 + Yv**2))
ax_inset.contourf(Xv, Yv, vortex, cmap='Reds')
ax_inset.set_title('Модель БКП')
ax_inset.axis('off')

plt.tight_layout()
plt.show()
```

```python
# РАСЧЕТЫ
print("Расчеты выбор от 1 до 3")
print("\n1. Гравитация на разных планетах")
print("2. Орбитальная скорость спутников")
print("3. Сравнение массы")

choice = input("\nВыберите расчет от 1 до 3): ")

if choice == '1':
    # гравитация
    m_str = input("Введите Ваш вес (в кг): ")
    try:
        m = float(m_str)
        planets_gravity = {
            'Меркурий': 3.7,
            'Венера': 8.87,
            'Земля': 9.81,
            'Марс': 3.71,
            'Юпитер': 24.79,
            'Сатурн': 10.44,
            'Уран': 8.87,
            'Нептун': 11.15
        }
        
        print("\nВаш вес на разных планетах:")
        for planet, g in planets_gravity.items():
            weight = m * g
            earth_weight = m * 9.81
            ratio = weight / earth_weight
            print(f"{planet:10} -> {weight:6.1f} Н (x{ratio:.2f})")
            
    except:
        print("Ошибка ввода")

elif choice == '2':
    # орбитальная скорость
    print("\nОрбитальная скорость на разных высотах над Юпитером:")
    altitudes = [1000, 10000, 421600]  # км (последнее - орбита Ио)
    
    for h in altitudes:
        r = (R_jupiter + h) * 1000  # перевод в метры
        v = np.sqrt(G * M_jupiter / r) / 1000  # км/с
        period = 2 * np.pi * r / (v * 1000) / 3600  # часы
        print(f"Высота {h:6} км: v = {v:.2f} км/с, T = {period:.1f} ч")

elif choice == '3':
    # сравнение массы
    earth_mass = 5.972e24  # кг
    jupiter_earth_ratio = M_jupiter / earth_mass


    print("\nМасса Юпитера составляет:")
    print(f"{jupiter_earth_ratio:.0f} масс Земли")
    
    # масса всех планет кроме Юпитера
    other_planets_mass = earth_mass * 400  # приблизительно
    total_mass = M_jupiter + other_planets_mass
    jupiter_percent = M_jupiter / total_mass * 100
    
    print(f"\nЮпитер содержит {jupiter_percent:.1f}% массы всех планет")
    print("(без учета Солнца)")

else:
    print("Неверный выбор")

# вывод
print("\n" + "="*50)
print("Юпитер - Ключевые характеристики:")
print(f"• Радиус: {R_jupiter:,} км ({R_jupiter/6371:.1f}× Земли)")
print(f"• Масса: {M_jupiter:.3e} кг ({M_jupiter/5.972e24:.0f}× Земли)")
print(f"• Ускорение свободного падения: {g_surface} м/с²")
print(f"• Период вращения: {T_rotation} часов")
print(f"• Температура облаков: {temp_cloud}°C")
print("="*50)
```
